```python
n_titles = len(list(usc.iter_titles()))
print(n_titles)

def apply_clustering(entities, distance_func, save_path=None):
    clustering = Clustering(entities, distance_func)
    print(clustering.get_clusters(n_titles))
    
    plt.figure(figsize=(200, 150))
    clustering.plot_dendrogram()
    plt.xticks(fontsize=14)
    if save_path:
        plt.savefig(save_path)
```
